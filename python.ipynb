```python
import os 
import urllib.request
import pandas as pd
import numpy as np
```

```python
DOWNLOAD_ROOT = "https://raw.githubusercontent.com/sebastiandiazmujica/digital57/master/data.csv"
os.makedirs("/content/data", exist_ok=True)
urllib.request.urlretrieve(DOWNLOAD_ROOT, "data.csv")
```

```python
data = pd.read_csv("data.csv", sep= ",", encoding='UTF-8')
data
```

```python
mayor_1240= data.PostalCode < 1240
data = data[mayor_1240]
data
```

```python
data['newCity'] = data['City'].apply(lambda x: x[-1])
data['newCountry'] = data['Country'].apply(lambda x: x[-1])
data['newCity'] = pd.to_numeric(data['newCity'])
data['newCountry'] = pd.to_numeric(data['newCountry'])
data['difference'] = data.newCity - data.newCountry
data = data[data.difference !=0 ]
data
```
